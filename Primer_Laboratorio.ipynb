## Introducción
Este laboratorio se trabaja con lectura y escritura de archivos en Python usando context managers, pandas DataFrames y pickle.

```python
## Ejercicio 1
Lectura y escritura de archivos TXT y CSV usando context managers.
```

```python
# Crear y escribir un archivo TXT
with open("ejemplo.txt", "w") as f:
    f.write("Hola, este es un archivo de texto hecho por Jimena.\n")
    f.write("Segunda línea.\n")
```

```python
# Leer el archivo TXT
with open("ejemplo.txt", "r") as f:
    contenido_archivo = f.read()

print(contenido_archivo)
```

```python
import csv

# Escribir CSV
with open("ejemplo.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["nombre", "edad"])
    writer.writerow(["Jimena", 20])
    writer.writerow(["Danilo", 19])
```

```python
# Leer CSV
with open("ejemplo.csv", "r") as f:
    reader = csv.reader(f)
    for fila in reader:
        print(fila)
```

 ### ¿En qué parte de un pipeline de datos se usan R y W?
La lectura se usa al inicio del pipeline, cuando se cargan los datos mientras que, la escritura se usa al final, cuando se guardan resultados, reportes o datos procesados.

## Ejercicio 3
Guardar y leer objetos con pickle.

```python
import pickle

datos = {
    "nombre": "Jimena",
    "edad": 20,
    "carrera": "Computer Science"
}

with open("datos.pkl", "wb") as f:
    pickle.dump(datos, f)
```

```python
# Leer el pickle
with open("datos.pkl", "rb") as f:
    datos_l = pickle.load(f)

print(datos_l)
```

### ¿Qué tipo de dato se puede almacenar con pickle?
Pickle puede almacenar objetos de Python, como listas, diccionarios, DataFrames, modelos y clases personalizadas, siempre que sean serializables.
